```python
#importamos las librerias que vamos a necesitar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#leemos el archivo html y las tablas GDP per capita
datos =pd.read_html('https://en.wikipedia.org/wiki/Economy_of_the_United_States', match ='GDP per capita')
#Numero total de tablas con ese mismo nombre
len(datos)
```

```python
datos[1]
```

```python
df=datos[1]
```

```python
list(df)
```

```python
#Eliminar columnas
df = df.drop(columns=['GDP (in Bil. US$PPP)','GDP per capita (in US$ PPP)'])
```

```python
df.info()
```

```python
df.columns =['Year','GDP','GDP_Per_Capita','GDP_growth','Inflation_Rate','Unemployment','Government_Debt']
```

```python
df.info()
```

```python
df['GDP_growth'] =df['GDP_growth'].replace({'%':''}, regex=True).astype('float')
```

```python
df.info()
```

```python
df['Inflation_Rate']=df['Inflation_Rate'].replace({'%':''}, regex=True).astype('float')
```

```python
df['Unemployment']=df['Unemployment'].replace({'%':''}, regex=True).astype('float')
```

```python
df['Government_Debt']=df['Government_Debt'].replace({'%':''}, regex=True).astype('float')
```

```python
df.info()
```

```python
df.Government_Debt.isnull().sum()
```

```python
df.to_csv('USA_economy.csv', index = False)
```
